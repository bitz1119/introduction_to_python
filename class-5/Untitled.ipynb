```python
import requests
inp = input("Enter what you want to search ").split()
url = "https://in.search.yahoo.com/search?p="
inp = '+'.join(inp)
url = url + inp 
print(url)
r = requests.get(url)
print(r)
```

```python
from bs4 import BeautifulSoup as bs
```

```python
s = bs(r.text,"html.parser")
```

```python
print(s.prettify())
```

```python
x = s.find("ol",{"class":"mb-15 reg searchCenterMiddle"})
```

```python
x
```

```python
y = x.find_all("li")
```

```python
for i in y:
    obj = i.find("h3") 
    if(obj != None):
        obj1 = obj.find("a")
        print(obj1.get_text(),obj1.get("href"))
```
